## 22.1 - ODE Initial Value Problem Statement ##

*   Differential Equation: Relationship between a function, its independent variable, and any number of its derivatives.
*   Ordinary Differential Equation (ODE): A differential equation where the indfependent variable, and the derivatives, is in one dimension.
*   nth order ODE: (d^n f(x)) / dx^n
*   Partial Differential Equation (PDE): A general form differential equation where the independent variable is a vector containing (x1, x2, ..., xm) and the partial derivatives can be of any order and with respect to any combination of variables.
*   General Solution: A g(x) that satisfies the differential equation
*   Particular Solution: A p(x) that satisfies the differential equation and n explicitly known valies of the solution, or its derivatives, at certain points. Also referrerd to as the solution.
*   Initial Value: A common set of known values for an ODE solution.
*   Initial Value Problem: Finding a solution to an ODE given an initial value.

## 22.2 - Reduction of Order ##
*   Of the methods designed to solve initial value problems, many are tailored to first order differential equations. This can be a problem when dealing with higher order differential equations, therefore it becomes prudent to reduce the order of a particular ODE to the first order.
*   State of a System: A collection of all relevant dependent variables
*   We can take a derivative of the state of a system to reduce the order of a differential equation to the first.
*   Linear ODE: An ODE that can be written in matrix form.

## 22.3 - The Euler Method ##
*   The Explicit Euler Formula allows the computation of an approximation for the state at S(tj+1) given the state at S(tj) using S(tj+1)=S(tj)+hF(tj,S(tj)).
*   Given an initial value for S0, the formula can be used to integrate the states up to S(tf).
*   Those values can then be used as an approximation of the solution to the differential equation
*   While this can get tedious and is error prone to perform by hand, there exists functionality in Python to achieve such a thing:

```python
# An example of using Python to approximate the solution to a differential equation using the Euler Method

import numpy as np
import matplotlib.pyplot as plt

# Define parameters
f = lambda t, s: np.exp(-t) # ODE
h = 0.1                     # Step size
t = np.arange(0, 1 + h, h)  # Numerical grid
s0 = -1                     # Initial Condition

# Explicit Euler Method
s = np.zeros(len(t))
s[0] = s0

for i in range(0, len(t) - 1):
    s[i + 1] = s[i] + h*f(t[i], s[i])

plt.figure(figsize = (12, 8))
plt.plot(t, s, 'bo--', label='Approximate') ; plt.plot(t, -np.exp(-t), 'g', label='Exact')
plt.title('Approximate and Exact Solution for Simple ODE with Step Size of 0.1')
plt.xlabel('t') ; plt.ylabel('f(t)')
plt.grid()      ; plt.legend(loc='lower right')
plt.show()
```

While the approximation is close, we can see that it can be even better given a smaller step size:
