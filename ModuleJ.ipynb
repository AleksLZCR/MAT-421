Graph Models:
*   Nodes: Objects modeled in a graph
  *   Degree: Number of edges connected to a node
*   Edges: Connections between nodes
*   Graph: G(V, E)
  *   V = the set of nodes
  *   E = the set of edges
  *   Directed: edges have direction
*   Path: Sequence of edges where nodes and edges are distinct
  *   Closed Path: Cycle
  *   Length: Number of edges traversed
*   N-hop neighborhood: The set of nodes within n hops distance from a node v
*   Python has the 'networkx' library that enables easy plotting of graphs:

```python
# An example of using the networkx library to plot a simple, directed graph

import networkx as nx
import matplotlib.pyplot as plt

# Create a directed graph
G = nx.DiGraph()

# Add nodes
G.add_node(1)
G.add_node(2)
G.add_node(3)
G.add_node(4)
G.add_node(5)

# Add edges
G.add_edge(1, 2)
G.add_edge(1, 3)
G.add_edge(2, 3)
G.add_edge(3, 4)
G.add_edge(5, 3)
G.add_edge(1, 5)
G.add_edge(4, 5)

# Plot
pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True)
plt.title("Directed Graph")
plt.show()
```

Laplacian Matrix:
*   An nxn adjacency matrix can be used to represent a graph with n nodes
  *   1 at row i col j: connection between vi, vj
  *   0 at row i col j: no connection between vi, vj
*   A Laplacian matrix encodes properties of a graphs structure
  *   Diagonal elements = degree of each node
  *   Off diagonal elements = connections between nodes
*   Has the following properties:
  *   L = D − M, where M is the adjacency matrix and D is the diagonal degree
matrix with Dii = ∑k Eik
  *   L is symmetric, positive, semi-definite. All eigenvalues are real and
non-negative, and L has a full set of n real and orthogonal eigenvectors
  *   If the graph G has c connected components, then L has c eigenvalues that
is 0
  *   For any vector x, x^T Lx = ∑{i, j}∈E Eij(xi − xj)^2
  *   For any vector x and scalars α, β , (αx +βe)^T * L(αx +β e) = α^2(x^T)Lx

## 4.2 - Spectral Graph Bipartitioning
*   Cut: the total number of edges between two disjoint sets of nodes
*   The goal of graph partitioning is to find a partition such that the cut is minimized
*   Two variants of performing this are the ratio cut and normalized cut




```python
# An example showing a cut in Python

import networkx as nx
import matplotlib.pyplot as plt

# Create graph
G = nx.Graph()
# Add nodes
G.add_node(1)
G.add_node(2)
G.add_node(3)
G.add_node(4)
G.add_node(5)

# Add edges
G.add_edge(1, 2)
G.add_edge(1, 3)
G.add_edge(2, 3)
G.add_edge(3, 4)
G.add_edge(5, 3)
G.add_edge(1, 5)
G.add_edge(4, 5)

# Define cut
cut_edges = [(1, 2), (1, 3), (2, 3)]

# remove cut edges
G_cut = G.copy()
G_cut.remove_edges_from(cut_edges)

# Plot the original graph
plt.figure(figsize=(8, 4))
plt.subplot(1, 2, 1)
pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True)
plt.title("Original Graph")

# Plot the cut
plt.subplot(1, 2, 2)
nx.draw(G_cut, pos, with_labels=True)
plt.title("Graph after Cut")

plt.tight_layout()
plt.show()
```
