## 2.2 - Probability Distribution ##
Probability:
*   The study of uncertainty
*   Used to quantify uncertainty in data analysis and prediction
*   Essential for modeling and understanting various model performance

Probability Distribution:
*   Mathematical function that gives the probabilities of occurrence of different possible outcomes for an experiment

Definitions:
*   Sample Space: Denoted by S; the set of all possible outcomes of that experiment
*   Event: Any collection of outcomes contained in S
*   Simple Event: An event that consists of exactly one outcome
*   Compound Event: An event that consists of more than one outcome
*   Probability Distribution: A function which assigns a number to each event that gives a precise measurement of how likely the event is to occur
*   A probability should not be greater than one, nor less than zero

Conditional Probability:
*   The likelihood of an event or outcome occurring based on the occurrence of a previous event or outcome
*   Expressed as a ratio of unconditional probabilities
*   Can be used to determine if one event occurring has any influence on another event occurring

```python
# An example using Python to determine if someone being wealthy has any bearing on them purchasing a luxury vehicle

# Define probabilities
prob_wealth = 0.3
prob_poor   = 1 - prob_wealth

prob_wealth_buy_lux = 0.6
prob_poor_buy_lux   = 0.1

# Calculate the total probability of buying a luxury car
total_prob_lux = (prob_wealth * prob_wealth_buy_lux) + (prob_poor * prob_poor_buy_lux)

# Calculate conditional probability of someone buying a luxury car given wealthy
cond_prob = (prob_wealth * prob_wealth_buy_lux) / total_prob_lux

print("Conditional Probability of buying a luxury car given wealthy:", cond_prob)
```

Discrete Random Variable:
*   Random Variable: any rule that associates a number with each outcome
*   Discrete Random Variable: A random variable whose possible values are a finite set or an infinite sequence

Cumulative Distribution Function:
*    Describes the probability that a random variable X takes on a value less than or equal to a given value x

Poisson Distribution:
*   Describes the number of events occurring in a fixed interval of time given a known average rate of occurrence and independence between events.

Python has libraries for several of the popular probability distributions such as poisson and cdf

```python
# Extending previous example to include probability distributions and graphs

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

# Define probabilities
prob_wealth = 0.3
prob_poor   = 1 - prob_wealth

prob_wealth_buy_lux = 0.6
prob_poor_buy_lux   = 0.1

# Calculate the total probability of buying a luxury car
total_prob_lux = (prob_wealth * prob_wealth_buy_lux) + (prob_poor * prob_poor_buy_lux)

# Calculate conditional probability of someone buying a luxury car given wealthy
cond_prob = (prob_wealth * prob_wealth_buy_lux) / total_prob_lux

# Generate a sample of car purchases over a period of time (let's say 100 time units)
time_units = 100
mean_purchase_rate = 10  # Average number of car purchases per time unit

# Generate a Poisson distribution for car purchases
poisson_data = np.random.poisson(mean_purchase_rate, time_units)

# Calculate cumulative probabilities for different levels of wealth
wealth_levels = np.linspace(0, 1, 100)  # Assume wealth levels vary from 0 to 1
cumulative_probs = []
for wealth_level in wealth_levels:
    prob_luxury_given_wealth_level = (wealth_level * prob_wealth_buy_lux) + ((1 - wealth_level) * prob_poor_buy_lux)
    cumulative_probs.append(prob_luxury_given_wealth_level)

plt.figure(figsize=(12, 6))

# CDF
plt.subplot(1, 2, 1)
plt.plot(wealth_levels, cumulative_probs, label='CDF of Buying Luxury Car')
plt.title('Cumulative Distribution Function')
plt.xlabel('Wealth Level')
plt.ylabel('Probability')
plt.legend()

# Poisson
plt.subplot(1, 2, 2)
plt.hist(poisson_data, bins=20, density=True, alpha=0.6, color='r', label='Poisson Distribution')
plt.title('Poisson Distribution of Car Purchases')
plt.xlabel('Number of Car Purchases')
plt.ylabel('Probability Density')
plt.legend()

plt.tight_layout()
plt.show()
```

Continuous Random Variable:
*   A random variable whose values comprise either a single interval or a union of disjoint intervals
*   Used for probability density functions (PDF)
*   Expected Value: Also referred to as the mean; the average value that one would expect the variable to take over a number of trials
*   Variance: The spread of the distribution; indicates how the actual values differ from the expected values

Normal Distribution:
*   Used in natural and social sciences
*   Represents real-valued random variables whose distributions are unknown
*   Standard Normal Distribution: Normal distribution with variance = 0 and standard distribution = 1

## 2.3 - Independent Variables and Random Samples ##

Joint Probability
*   The probability of two or more events happening together
*   This requires considering two discrete or continuous random variables
*   Marginal Distribution: The probability distribution of the variables contained in the subset without reference to the values of the other variable

Independent Random Variables
*   Two variables, X and Y, are said to be independent if
*   p(x, y) = pX (x)  pY (y) when X and Y are discrete or
*   f (x, y) = fX (x)  fY (y) when X and Y are continuous.
*   Otherwise, X and Y are dependent

Correlation
*   Can indicate a predictive relationship
*   Covariance: measure of the joint variability of two random variables



```python
# An example of plotting positive covariance using python

import numpy as np
import matplotlib.pyplot as plt

# Generate Data
x = np.random.normal(0, 1, 100)          # random data for variable x
y = 2 * x + np.random.normal(0, 1, 100)  # random data for variable y -- positive covariance with x

plt.scatter(x, y)
plt.title('Positive Covariance')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid(True)
plt.show()
```

Random Samples
*   A randomly selected subset of a population
*   Often used in practice
*   Random variables X1, X2, ... are said to form a random sample if
*   The Xs are independent random variables and every X has the same probability distribution

Central Limit Theorem
*   The properly normalized sum of independent random variables tends toward a normal distribution even if the original variables aren't normally distributed
*   Increases the applications of statistical methods that work for normal distributions.
