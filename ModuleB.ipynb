17.1 Interpolation Problem Statement



*   Interpolation is the process of determining a function given a set of data points
*   It differs from regressions as regressions assumes measurement error exists. Interpolation will result in a function that traverses all data points.



Given a set of independent values xi and dependent values yi, interpolation can be used to find an estimation function y such that y(xi) = yi

```python
import matplotlib.pyplot as plt
x = [1, 2, 3, 4]
y = [2, 4, 3.5, 4.5]

plt.figure(figsize = (3,3))
plt.plot(x,y,'o')
plt.title('Data Set')
plt.ylim(0, 10)
plt.xlabel('xi')
plt.ylabel('yi')
plt.show()
```

The advantage as compared to regressions is that the user does not need an underlying model for the data which can be complex to derive. The data must be accurate for this approach though.

17.2 Linear Interpolation



*   Linear interpolation assumes that an estimated point lies along the line joining the nearest left and right points.
*   The equation for this is as follows:

  y(x) = yi + [(yi+1 - yi)(x - xi)] / [xi+1 - xi]



```python
# An example of finding the linear interpolation at x = 4.1 on the data set x = [0, 1, 2, 3, 4, 5] y = [1, 3, 2, 7, 0, 1]
from scipy.interpolate import interp1d
import matplotlib.pyplot as plt

x = [0, 1, 2, 3, 4, 5]
y = [1, 3, 2, 7, 0, 1]

f = interp1d(x, y)
y_hat = f(4.1)
print(y_hat)

plt.figure(figsize = (7,7))
plt.plot(x, y, '-ob')
plt.plot(4.1, y_hat, 'ro')
plt.title('Linear Interpolation at x = 4.1')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

17.3 Cubic Spline Interpolation



*   Rather than one function as in linear regression, cublic spline interpolation is a set of piecewise cubic functions.
*   For n points, there are n-1 cubic functions to find; each function requires 4 coefficients.
*   Two final equations are used to establish boundary conditions at the ends of the interpolation.



```python
# An example of cubic spline interpolation over the data set x = [0, 1, 2, 3, 4, 5] and y = [1, 3, 1, 6, 1, 3] for 0 <= x <= 5

from scipy.interpolate import CubicSpline
import matplotlib.pyplot as plt
import numpy as np

x = [0, 1, 2, 3, 4, 5]
y = [1, 3, 1, 6, 1, 3]

f = CubicSpline(x, y, bc_type='natural') # using natural for coefficients
x_new = np.linspace(0, 6, 100)
y_new = f(x_new)

plt.figure(figsize = (8,8))
plt.plot(x_new, y_new, 'b') # cubic interpolation
plt.plot(x, y, 'ro')        # overlay points
plt.title('Cubic Spline Interpolation')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

This type of interpolation would more accurately reflect things like  velocities which cannot instantaneously change slope
